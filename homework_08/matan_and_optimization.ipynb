$$
f(x_1, x_2) = x^3_1−2x_1x_2 + x^2_2−3x_1−2x_2, \ x ∈R^2
$$

Частная производная по $x_1$:
$$
\frac{\partial f}{\partial x_1} = 3x_1^2 - 2x_2 - 3
$$

Частная производная по $x_2$:
$$
\frac{\partial f}{\partial x_2} = -2x_1 + 2x_2 - 2
$$

Найдите критические точки $x_c$, такие что $\nabla f (x_c) = 0$.
$$
\frac{\partial f}{\partial x_1} = 0 ; \frac{\partial f}{\partial x_2} = 0
$$

Решим систему уравнений, прировняв частные производные к нулю

$$
\begin{align}
    3x_1^2 - 2x_2 - 3 = 0 \quad \text{(1)} \\
    -2x_1 + 2x_2 - 2 = 0 \quad \text{(2)}
\end{align}
$$

Из 2ого уравнения выразим $x_2$:

$$
\begin{align}
    2x_2 = 2 + 2x_1 \implies x_2 = 1 + x_1
\end{align}
$$

Подставим это выражение в 1ое уравнение:

$$
\begin{align}
    3x_1^2 - 2(1 + x_1) - 3 = 0 \implies 3x_1^2 - 2 - 2x_1 - 3 = 0 \implies 3x_1^2 - 2x_1 - 5 = 0
\end{align}
$$

Решим квадратное уравнение:

$$
\begin{align}
    D = b^2 - 4ac \implies (-2)^2 - 4 \cdot 3 \cdot (-5) = 4 + 60 = 64 \\
    x_1 = \frac{10}{6} = \frac{5}{3} \quad \text{и} \quad x_1 = \frac{-6}{6} = -1
\end{align}
$$

Подставим найденные значения $x_1$ обратно в выражение для $x_2$:

Для $x_1 = \frac{5}{3}: \ x_2 = 1 + \frac{5}{3} = \frac{8}{3}$

Для $x_1 = -1: \ x_2 = 1 + (-1) = 0$

Получились 2 критические точки:

$\left( \frac{5}{3}, \frac{8}{3} \right)$

$(-1, 0)$

2. **Проверьте, что функция** $\ f(x_1, x_2) = \ln(\sqrt{x_1} + \sqrt{x_2}) \$ **удовлетворяет уравнению:**

$$
x_1\frac{\partial f}{\partial x_1} + x_2\frac{\partial f}{\partial x_2} = \frac{1}{2}
$$

частные производные для $x_1 \ и\ x-2$:
$$
\frac{\partial f}{\partial x_1} = \frac{1}{\sqrt{x_1} + \sqrt{x_2}} \cdot \frac{1}{2\sqrt{x_1}} = \frac{1}{2\sqrt{x_1}(\sqrt{x_1} + \sqrt{x_2})}
$$

$$
\frac{\partial f}{\partial x_2} = \frac{1}{\sqrt{x_1} + \sqrt{x_2}} \cdot \frac{1}{2\sqrt{x_2}} = \frac{1}{2\sqrt{x_2}(\sqrt{x_1} + \sqrt{x_2})}
$$

Подставим в уравнение
$$
\begin{align}
x_1 \cdot \frac{1}{2\sqrt{x_1}(\sqrt{x_1} + \sqrt{x_2})} + x_2 \cdot \frac{1}{2\sqrt{x_2}(\sqrt{x_1} + \sqrt{x_2})} = \frac{1}{2} \implies \\
\frac{x_1}{2\sqrt{x_1}(\sqrt{x_1} + \sqrt{x_2})} + \frac{x_2}{2\sqrt{x_2}(\sqrt{x_1} + \sqrt{x_2})} = \frac{1}{2} \implies \\
\frac{\sqrt{x_1}}{2(\sqrt{x_1} + \sqrt{x_2})} + \frac{\sqrt{x_2}}{2(\sqrt{x_1} + \sqrt{x_2})} \implies \\
\frac{\cancel{\sqrt{x_1} + \sqrt{x_2}}}{2\cancel{(\sqrt{x_1} + \sqrt{x_2})}} = \frac{1}{2}
\end{align}
$$

Это подтверждает, что функция $f(x_1, x_2) = \ln(\sqrt{x_1} + \sqrt{x_2})$ удовлетворяет данному уравнению

4. **(Куб Евклидовой нормы). Найти первый дифференциал**  $df(x)$, а также градиент $\nabla f(x)$ функции:

$$
f(x) = \frac{1}{3} \|x\|_2^3 \quad x \in \mathbb{R}^n \setminus \{0\}
$$

Вводим новую переменную:
$$r = \|x\|_2$$

Тогда:
$$f(x) = g(r) = \frac{1}{3} r^3 \text{, а } \ r(x) = \|x\|_2$$


Находим дифференциал функции $g(r)$:

Сначала находим производную g(r): $g'(r)=r^2$

Находим дифференциал $dg(r)$: $dg(r)=g'(r)dr=r^2dr$

Находим дифференциал $r = \|x\|_2 = (x,x)^\frac{1}{2}$:
$$
dr = d((x,x)^\frac{1}{2}) = \frac{1}{2}(x,x)^{-\frac{1}{2}} \cdot d(x,x) = \frac{1}{2}(x,x)^{-\frac{1}{2}} \cdot 2(x, dx) = \frac{1}{2}r \cdot 2(x, dx) = \frac{(x, dx)}{r}
$$

дифференциал $dg(r)$:
$$
dg(r) = r^2 \cdot dr = r^2 \cdot \frac{(x, dx)}{r} = r(x, dx) = \|x\|_2(x, dx)
$$

Дифференциал функции f(x):

$$
f(x) = g(x) \implies f(x) = r(x, dx) = \|x\|_2(x, dx)
$$

Градиент функции $\nabla f(x)$:

$$
f(x)=g(r) \implies \nabla f(x) = g'(r) \cdot \nabla r
$$

где $g'(r) = r^2$, находим $\nabla r = \frac{\partial r}{\partial x_i}$:

$r = (x,x)^\frac{1}{2}$, пусть $s = (x,x)$, тогда $r = s^\frac{1}{2}$

Находим частную производную r по s:
$$
\frac{\partial r}{\partial s} = \frac{1}{2}s^{-\frac{1}{2}} = \frac{1}{2} \cdot \frac{1}{r}
$$

Находим частную производную S по $x_i$:
$$
\frac{\partial s}{\partial x_i} = 2x_i
$$

Упростим:
$$
\nabla r = \frac{\partial r}{\partial x_i} = \frac{x}{r}
$$

Градиент функции $\nabla f(x)$:

$$
\nabla f(x) = g'(r) \cdot \nabla r = r^2 \cdot \frac{x}{r} = rx = \|x\|_2x
$$

5. **(Квадрат Евклидовой нормы).** Найдите первый дифференциал $df(x)$, и градиент $\nabla f(x)$ функции:

$$
f(x) = \|Ax\|_2^2 \quad x \in \mathbb{R}^n \setminus \{0\} \quad A \in \mathbb{R}^{m \times n}
$$

Запишем функцию в явном виде:
$$
|Ax|_2 = \sqrt{(Ax)^T (Ax)} \implies |Ax|_2^2 = (Ax)^T (Ax)
$$

Тогда

$$
f(x) = |Ax|_2^2 = (Ax)^T (Ax) = x^T A^T A x
$$

выделим переменную $B = A^T A$:

$$
f(x) = x^T A^T A x = x^T B x
$$

Найдем дифференциал $df(x)$:

$$
df(x) = (B x)^T dx + x^T B dx = 2 x^T B dx \implies df(x) = 2 (Ax)^T A dx
$$

Градиент $\nabla f(x)$:

$$
df(x) = 2 (Ax)^T Adx \implies df(x) = \nabla f(x)^T dx \implies \nabla f(x)^T = 2 (Ax)^T A \implies \nabla f(x) = 2 A^T Ax
$$
