**Линейная алгебра**


**1. Даны матрицы:**
$$
A = \begin{bmatrix}
2 & -4 \\
3 & 5 \\
-1 & 0
\end{bmatrix}, \quad
B = \begin{bmatrix}
1 & 2 & 7 \\
-3 & -4 & 0 \\
5 & 2 & 1
\end{bmatrix}, \quad
C = \begin{bmatrix}
6 & -3 & 9 \\
4 & -5 & 2 \\
8 & 1 & 5
\end{bmatrix}
$$

Посчитайте матрицу $ D = A^\top C - 2 A^\top B^\top $. Приведите полную последовательность вычислений.


**Решение**

Найдем транспонированные матрицы

Матрица А
$$
A = \begin{bmatrix}
2 & -4 \\
3 & 5 \\
-1 & 0
\end{bmatrix}
\implies
A^\top = \begin{bmatrix}
2 & 3 & -1 \\
-4 & 5 & 0
\end{bmatrix}
$$

Матрица B
$$
B = \begin{bmatrix}
1 & 2 & 7 \\
-3 & -4 & 0 \\
5 & 2 & 1
\end{bmatrix}
\implies
B^\top = \begin{bmatrix}
1 & -3 & 5 \\
2 & -4 & 2 \\
7 & 0 & 1
\end{bmatrix}
$$

Вычислим $( A^\top C )$:

\begin{align}
    A^\top C &=
    \begin{bmatrix}
    2 & 3 & -1 \\
    -4 & 5 & 0
    \end{bmatrix}
    \begin{bmatrix}
    6 & -3 & 9 \\
    4 & -5 & 2 \\
    8 & 1 & 5
    \end{bmatrix}\\
    &= \begin{bmatrix}
    [ 2 \cdot 6 + 3 \cdot 4 + (-1) \cdot 8 = 16 ] &
    [ 2 \cdot (-3) + 3 \cdot (-5) + (-1) \cdot 1 = -22 ] &
    [ 2 \cdot 9 + 3 \cdot 2 + (-1) \cdot 5 = 18 + 6 - 5 = 19 ] \\
    [-4 \cdot 6 + 5 \cdot 4 + 0 \cdot 8 = -4 ] &
    [-4 \cdot (-3) + 5 \cdot (-5) + 0 \cdot 1 = -13 ] &
    [-4 \cdot 9 + 5 \cdot 2 + 0 \cdot 5 = -26 ]
    \end{bmatrix}\\
    &= \begin{bmatrix}
    16 & -22 & 19 \\
    -4 & -13 & -26
    \end{bmatrix}
\end{align}

Вычислим $( A^\top B^\top )$:

\begin{align}
    A^\top B^\top &=
    \begin{bmatrix}
    2 & 3 & -1 \\
    -4 & 5 & 0
    \end{bmatrix}
    \begin{bmatrix}
    1 & -3 & 5 \\
    2 & -4 & 2 \\
    7 & 0 & 1
    \end{bmatrix} \\
    &= \begin{bmatrix}
    [ 2 \cdot 1 + 3 \cdot 2 + (-1) \cdot 7 = 1 ] & [ 2 \cdot (-3) + 3 \cdot (-4) + (-1) \cdot 0 = -18 ] & [ 2 \cdot 5 + 3 \cdot 2 + (-1) \cdot 1 = 15 ] \\
    [ -4 \cdot 1 + 5 \cdot 2 + 0 \cdot 7 = 6 ] & [ -4 \cdot (-3) + 5 \cdot (-4) + 0 \cdot 0 = -8 ] & [ -4 \cdot 5 + 5 \cdot 2 + 0 \cdot 1 = -10 ]
    \end{bmatrix} \\
    &= \begin{bmatrix}
    1 & -18 & 15 \\
    6 & -8 & -10
    \end{bmatrix}
\end{align}

Вычислим $( 2 A^\top B^\top )$:

$$
2 A^\top B^\top =
2 \begin{bmatrix}
1 & -18 & 15 \\
6 & -8 & -10
\end{bmatrix}
=
\begin{bmatrix}
2 & -36 & 30 \\
12 & -16 & -20
\end{bmatrix}
$$

Вычислим $( D = A^\top C - 2 A^\top B^\top )$:


\begin{align}
    D &= \begin{bmatrix}
    16 & -22 & 19 \\
    -4 & -13 & -26
    \end{bmatrix} -
    \begin{bmatrix}
    2 & -36 & 30 \\
    12 & -16 & -20
    \end{bmatrix} \\
    &= \begin{bmatrix}
    [ 16 - 2 = 14 ] & [ -22 + 36 = 14 ] & [ 19 - 30 = -11 ] \\
    [ -4 - 12 = -16 ] & [ -13 + 16 = 3 ] & [ -26 + 20 = -6 ]
    \end{bmatrix} \\
    &= \begin{bmatrix}
    14 & 14 & -11 \\
    -16 & 3 & -6
    \end{bmatrix}
\end{align}

Ответ

$$
D = \begin{bmatrix}
14 & 14 & -11 \\
-16 & 3 & -6
\end{bmatrix}
$$

2. **Дано выражение:**

$$
3 \cdot \begin{bmatrix}
x & 2 & 3 \\
-1 & y & 4
\end{bmatrix}
+ 2 \cdot \begin{bmatrix}
1 & 2 & -5 \\
2 & -6 & z
\end{bmatrix}
= \begin{bmatrix}
8 & v & -1 \\
1 & 6 & 4
\end{bmatrix}
$$

Найдите значения $x, y, z \ и\ v$, при которых выражение верно.

**Решение**

\begin{align*}
    3 \cdot \begin{bmatrix}
    x & 2 & 3 \\
    -1 & y & 4
    \end{bmatrix}
    + 2 \cdot \begin{bmatrix}
    1 & 2 & -5 \\
    2 & -6 & z
    \end{bmatrix}
    = \begin{bmatrix}
    8 & v & -1 \\
    1 & 6 & 4
    \end{bmatrix}
    \implies \\
    \implies
    \begin{bmatrix}
    3x & 6 & 9 \\
    -3 & 3y & 12
    \end{bmatrix}
    + \begin{bmatrix}
    2 & 4 & -10 \\
    4 & -12 & 2z
    \end{bmatrix}
    = \begin{bmatrix}
    8 & v & -1 \\
    1 & 6 & 4
    \end{bmatrix}
    \implies \\
    \implies
    \begin{bmatrix}
    3x + 2 & 10 & -1 \\
    1 & 3y - 12 & 12 + 2z
    \end{bmatrix}
    = \begin{bmatrix}
    8 & v & -1 \\
    1 & 6 & 4
    \end{bmatrix}
    \implies \\
    \implies
    \begin{bmatrix}
    3x + 2 = 8 & 10 = v & -1 = -1 \\
    1 = 1  & 3y - 12 = 6 & 12 + 2z = 4
    \end{bmatrix}
    \implies \
    \begin{bmatrix}
    x = 2 & 10 = v & -1 \\
    1  & y = 6 & z = -4
    \end{bmatrix}
\end{align*}

**Ответ**

$$
\begin{align}
x = 2 \\
y = 6 \\
z = -4 \\
v = 10
\end{align}
$$

**3. Задание**

Относительно канонического (стандартного) базиса в R2 даны три вектора
$$
\mathbf{a_1} = \begin{pmatrix} 2 \ -5 \end{pmatrix}^T, \\
\mathbf{a_2} = \begin{pmatrix} -1 \ 3 \end{pmatrix}^T, \\
\mathbf{x} = \begin{pmatrix} 1 \ -4 \end{pmatrix}^T
$$
Примите векторы a1, a2 за новый базис B, предварительно проверив, что они линейно
независимы.
* (a) Найдите координаты $[x]_B$ вектора x в новом базисе.
* (b) Предположим, что координаты вектора y в базисе B заданы $[y]_B = \begin{pmatrix} 1 \ 1 \end{pmatrix}^T$ . Найдите координаты
вектора y в стандартном базисе.



**Решение**

* Проверка линейной независимости.
    Для проверки линейной независимости, представим матрицу $А$ составленной из векторов $\mathbf{a_1} \ и \ \mathbf{a_2}$ и найдем ее определитель:

    $$
    A = \begin{bmatrix} 2 & -1 \\
    -5 & 3 \end{bmatrix}
    \implies
    (2)(3) - (-1)(-5) = 6 - 5  = 1
    $$

    Определитель не равен 0, это означает что векторы $\mathbf{a_1} \ и \ \mathbf{a_2}$ линейно независимы

* Нахождение координат вектора $\mathbf{x}$ в новом базисе $B$

    Подставляем векторы в уравнение:

    $$
    \begin{bmatrix}
    1 \\ -4
    \end{bmatrix}
    = c_1 \begin{bmatrix}
    2 \\ -5
    \end{bmatrix}
    + c_2 \begin{bmatrix}
    -1 \\ 3
    \end{bmatrix}
    $$

    Делим уравнение на 2 компоненты:
    $$
    \begin{align*}
    1 = 2c_1 - c_2 \quad \text{(1)} \\
    -4 = -5c_1 + 3c_2 \quad \text{(2)}
    \end{align*}
    $$

    Решим систему уравнений:

    * Выразим c_2 из уравнения (1):
        $$c_2 = 2c_1 - 1$$
    * Подставим выражение в уравнение (2):
        $$
        \begin{align*}
        -4 = -5c_1 + 3(2c_1 - 1) \implies \\
        -4 = -5c_1 + 6c_1 - 3 \implies \\
        -4 + 3 = c_1 \implies \\
        c_1 = -1
        \end{align*}
        $$
    * Подставим в уравнение для $с_2$:
        $$
        \begin{align*}
        c_2 = 2(-1) - 1 \implies \\
        c_2 = -3
        \end{align*}
        $$

    **Результат**: координаты вектора $\mathbf{x}$ в новом базисе $B$:
        $[\mathbf{x}]_B = \begin{bmatrix} -1 \\ -3 \end{bmatrix}$

* Нахождение координат вектора $\mathbf{y}$ в стандартном базисе

    Вектор $\mathbf{y}$ можно выразить как линейную комбинацию векторов базиса: \
    $\mathbf{y} = c_1 \mathbf{a_1} + c_2 \mathbf{a_2}$ \
    где $c_1$ и $c_2$ — это координаты вектора $\mathbf{y}$ в базисе $B$ \
    подставим значения $c_1$ и $c_2$ в уравнение:

    $$
    \mathbf{y} = 1 \cdot \mathbf{a_1} + 1 \cdot \mathbf{a_2} \implies \\
    \mathbf{y} = 1 \cdot
    \begin{bmatrix} 2 \\ -5 \end{bmatrix}
    + 1 \cdot \begin{bmatrix} -1 \\ 3 \end{bmatrix}
    \implies \\
    \mathbf{y} = \begin{bmatrix} 2 \\ -5 \end{bmatrix} +
    \begin{bmatrix} -1 \\ 3 \end{bmatrix} =
    \begin{bmatrix} 2 - 1 \\ -5 + 3 \end{bmatrix} =
    \begin{bmatrix} 1 \\ -2 \end{bmatrix}
    $$

    **Результат**: координаты вектора $\mathbf{y}$ в стандартном базисе:
        $\mathbf{y} = \begin{bmatrix} 1 \\ -2 \end{bmatrix}$

**Задание 4**

* Сгенерируйте случайную квадратную матрицу A ∈Rn×n, n ≥100.
* Выполните сингулярное разложение этой матрицы, и получите три матрицы: U , S, V ⊤
* Выполняйте аппроксимацию матрицы A с рангом r, меняя его значение

* и каждый раз считайте ошибку апроксимации (как восстановленная матрица отличается от исходной)

* постройте график зависимости ошибки аппроксимации матрицы от ранга r

```python
import numpy as np
import matplotlib.pyplot as plt

n = 100
A = np.random.rand(n, n)

U, S, VT = np.linalg.svd(A)

errors = []

for r in range(2, n + 1):
    S_r = np.zeros((n, n))
    np.fill_diagonal(S_r, S[:r])

    A_r = U[:, :r] @ S_r[:r, :r] @ VT[:r, :]

    error = np.linalg.norm(A - A_r, 'fro')
    errors.append(error)

plt.figure(figsize=(10, 6))
plt.plot(range(2, n + 1), errors, marker='o')
plt.title('Ошибка аппроксимации матрицы от ранга r')
plt.xlabel('Ранг r')
plt.ylabel('Ошибка аппроксимации (Frobenius норма)')
plt.grid()
plt.xticks(range(2, n + 1, 5))
plt.show()
```
